```python
df1.columns = df1.columns.map(lambda x: x.lower())

df2.columns = df2.columns.map(lambda x: x.lower())

#df1.columns
```

```python
#Let's create a derived variable for the unemployment scenario.
#It is a very crude way of measuring the unemployment ratio.

#unemp = tot_p/tot_work_p
```

```python
form1 = smf.wls(formula = "zz ~ ((tot_work_p/(tot_p - p_06))*(non_work_p/(tot_p - p_06))) + ((p_lit/(tot_p - p_06))*(p_ill/(tot_p - p_06)))  ", data = df1)

res1 = form1.fit()
print res1.summary()
```

```python
formula1 = smf.wls(formula = "zz_ ~ ((tot_work_p/(tot_p - p_06))*(non_work_p/(tot_p - p_06))) + ((p_lit/(tot_p - p_06))*(p_ill/(tot_p - p_06)))  ", data = df2)

result1 = formula1.fit()
print result1.summary()
```

```python
form2 = smf.wls(formula = "zz ~ p_06/tot_p + (p_lit/(tot_p - p_06))*(tot_work_p/(tot_p - p_06)) + (p_lit/(tot_p - p_06))*(main_ot_p/(tot_p - p_06)) +  (p_ill/(tot_p - p_06)) * (main_al_p/ (tot_p - p_06)) + (margwork_p) +  (marg_ot_p/main_ot_p) + (p_lit/(tot_p - p_06)) *  main_hh_p + p_sc*p_st", data = df1)

res2 = form2.fit()
print res2.summary()
```

```python
formula2 = smf.wls(formula = "zz_ ~ p_06/tot_p + (p_lit/(tot_p - p_06))*(tot_work_p/(tot_p - p_06)) + (p_lit/(tot_p - p_06))*(main_ot_p/(tot_p - p_06)) +  (p_ill/(tot_p - p_06)) * (main_al_p/ (tot_p - p_06)) + (margwork_p) +  (p_lit/(tot_p - p_06)) *  (marg_ot_p/main_ot_p) + (p_lit/(tot_p - p_06))*main_hh_p + p_sc*p_st", data = df2)

result2 = formula2.fit()
print result2.summary()
```

<h1> Inferences from the dataset </h1>

<h3>Some interesting observations</h3>
<ul>
<li> From the kds plots, it can be infered that the z-score of most of the districts have decreased, which means that some districts have dominated/had a very high migration rates than the others.</li>
<li> One potential outlier can be Ariyalur, which is a newly formed district, so it's population is pretty low in 1991 than 2001, as some more nearby towns have agglomerated to form the district.</li>
<li> Another outlier is Namakkal, which has a very interesting behavious. For the year 2001, it's z-score is less than the mean(I mean some good deviation too), and in 2011, it as again a pretty good positive deviation from the mean. A very interesting behaviour, reason being the increasing educational resources sprouting up in the area. </li>
<li>But, removing outliers means a significant loss in the data, as there are a very limited observations across two time-stamps.</li>
<li>While Chennai is a troublesome outlier in the population sstatistic (almost chnging the shape of data), it however, remains in the cluster around 0, when the z-score of migration rate is considered.</li>
<li> Caste plays a good role in influencing the migration rate. </li>
</ul>

<h3> Derived variables created </h3>

<ol>
<li> Proportion of under-aged population (just for keeping them away from dealing with the employment stats). </li>
<li> Ratio of total working population to the total +6 aged population; multiplied by the literacy ratio. </li>
<li> Ratio of total working population(agricultural) to the total +6 aged population; multiplied by the illiteracy ratio. </li> 
<li> literacy ratio multiplied with the ratio of working (others) population </li>
<li> A similar multiplication of the literacy ration with the marginal worker populations. </li>
</ol>
