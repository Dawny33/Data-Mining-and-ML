<h3> But, is poverty really decreasing at a rate the data is depicting? </h3>

<p> There is a danger of the poverty count again raising, because the data suggests that there is a steep decline in the agricultural employment, which in turn contributes to a steep increase in the type-2 and type-3 employment. So, from this, it can be infered that due to mechanization of agricultural equipment, people are migrating from the rural areas to the urban areas due to the decline of employment opportunities in the agricultural domain. So, this results in a flow of population from the rural areas to the urban areas. </p>

<p> But on the other hand, due to the increase of population, and in turn the population density, the employment opportunities are decreasing. The fact that about 65-67% (infered from World bank data) of the Indian Population consists of youth, the unemployment rate is bound to increase further. This <a href="https://github.com/Dawny33/Data-Mining-and-ML/blob/master/Intern%20Project/World_Poverty_Analysis/RBI_Data/RBI_Educational_Instis.xls" target="blank">data</a> supports the fact that the youth population in India is ever increasing, az the number of enrollments in educationsal institutions are steeply increasing. </p>

<p> As the data is keenly observed, the unemployed population per total population ratio is steadily increasing, which means that, for the years to come, the ratio is bound to increase. </p>

<p> This data on the unemployment rate of the respective India states gives a rough picture that, in an overall perspective, the unemployment rate has been decreasing. 
But on careful observation, it can be infered that the states, which lie in the top percentile with respect to the total population, have their unemployment rate increasing as opposed to the ones in the lower quantile
This phenomenon supports the above inference, which chalks out a relationship between the population of the state(%change) and the employment rate. </p>

<p>From the India census data of 2011, it can be infered that the districts of India's top-10 most populated states have a population density of atleast 1000 per sq km.
It is also interesting to note that the rural population of these states are about 70%, except Maharashtra, which is considerable due to presence of high-revenue generating cities like mumbai and Pune.
So, if this population increment continues, the population density would increase close to about 2000. Even though the sanitation facilities, water sources, etc. are improving, their increment however, cannot sustain the population growth rate.
So, with such a huge population density, the available resources per person would decrease, in the end giving rise to further poverty, keeping the dropping rates of employment and the ever-increasing migration rate from rural to the urban areas.
</p>
