```python
updated_f = f[(f.cancel_dateline > 0)]

#updated_f[['order_created_at', 'cancel_dateline']]

updated_f['age'] = (updated_f['cancel_dateline'] - updated_f['order_created_at']) *3.80265e-7

updated_f[['order_created_at', 'cancel_dateline', 'age']]
```

```python
updated_f[['age', 'licensed_seats']]
```

```python
x,y = updated_f['age'], updated_f['licensed_seats']

sns.jointplot(x, y, kind="hex", stat_func=kendalltau, color="#4CB391")
```

```python
import numpy as np
import numpy.random
import matplotlib
import matplotlib.pyplot as plt

N = 50
#updated_f = f[(f.cancel_dateline > 0)]


p = updated_f
x = p['licensed_seats']
y = p['age']
#colors = np.random.rand(N)

#area = np.pi * (15 * np.random.rand(N))**2 # 0 to 15 point radiuses

#fig = plt.scatter(x, y,s = 300,  c=p['licensed_seats'], alpha=0.5)   #s=area,

from matplotlib import pyplot as plt    

fig = plt.figure()

m, b = np.polyfit(x, y, 1)

plt.scatter(x,y,s = 300,  c=p['licensed_seats'], alpha=0.5)
fig.suptitle('Churn Plot for customers with >31 seats', fontsize=20)
fig.set_size_inches(18.5, 10.5)
plt.xlabel('Number of seats', fontsize=18)
plt.ylabel('Age of customer', fontsize=16)



#fig.savefig('test.jpg')
plt.show()
```
